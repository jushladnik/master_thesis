```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import pickle

ROOT_PATH = '/home/jhladnik/'
```

```python
def label_axes(fig):
    fig.axes[0].set_xlabel('news')
    fig.axes[0].xaxis.set_label_position('top') 
    fig.axes[0].set_ylabel('word2vec')

    fig.axes[1].set_xlabel('Twitter')
    fig.axes[1].xaxis.set_label_position('top') 
    fig.axes[1].set_ylabel('')

    fig.axes[2].set_xlabel('')
    fig.axes[2].set_ylabel('fastText')

    fig.axes[3].set_xlabel('')
    fig.axes[3].set_ylabel('')
    return fig
```

## default parameters

```python


%matplotlib inline

param_name = 'num_topics'
param_value = 15

with open(f"{ROOT_PATH}masters-thesis/results/plots/sim_matrix_{param_value}_{param_name}.pickle",'rb') as fid:
    fig = pickle.load(fid)

fig = label_axes(fig)

plt.suptitle(f"Topic model similarity matrices with default parameters")

plt.tight_layout()
plt.savefig(f"{ROOT_PATH}masters-thesis/results/plots/sim_matrix_{param_value}_{param_name}_labeled.png")
pickle.dump(fig, open(f"{ROOT_PATH}masters-thesis/results/plots/sim_matrix_{param_value}_{param_name}_labeled.pickle", 'wb')) 
plt.show()


plt.show()
```
